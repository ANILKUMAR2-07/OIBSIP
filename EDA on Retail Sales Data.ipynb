```python
import pandas as pd 
import seaborn as sns
import numpy as np 
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
df = pd.read_csv('Diwali Sales Data.csv', encoding='unicode_escape')
```

```python
df.info()
```

```python
df.drop(['Status','unnamed1'],axis=1, inplace=True)
```

```python
df.describe()
```

```python
df.columns
```

```python
ax = sns.countplot(x='Gender',data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gen = df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x='Gender', y='Amount', data=sales_gen)
```

```python
ax = sns.countplot(x='Age Group',data=df,hue='Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_age = df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x='Age Group', y='Amount', data=sales_age)
```

```python
sales_st = df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x='State', y='Orders', data=sales_st)
```

```python
sales_Amu = df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(15)
sns.barplot(x='State', y='Amount', data=sales_Amu)
```

```python
ax = sns.countplot(x='Marital_Status',data=df,hue='Gender')
sns.set(rc={'figure.figsize':(7,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_st = df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(x='Marital_Status', y='Amount', data=sales_st,hue='Gender')
```

```python
ax = sns.countplot(x='Occupation',data=df)
sns.set(rc={'figure.figsize':(20,10)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_st = df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)                                                                     
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x='Occupation',y='Amount', data=sales_st)
```

```python
ax = sns.countplot(x='Product_Category',data=df)
sns.set(rc={'figure.figsize':(20,10)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_st = df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)                                                                     
sns.set(rc={'figure.figsize':(20,10)})
sns.barplot(x='Product_Category',y='Amount', data=sales_st)
```

```python
sales_st = df.groupby(['Product_ID'],as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)                                                                     
sns.set(rc={'figure.figsize':(20,10)})
sns.barplot(x='Product_ID',y='Orders', data=sales_st)
```
